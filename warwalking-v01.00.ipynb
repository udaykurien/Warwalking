```python
import pandas as pd
import folium
import os
```

```python
def gen_filepath(fname, data_rel_path = 'data'):
    cwd_abs_path = os.path.abspath(os.getcwd())
    data_abs_path = os.path.join(cwd_abs_path, data_rel_path)
    data_file_abs_path = os.path.join (data_abs_path, fname)
    return data_file_abs_path

file_path = gen_filepath('WigleWifi_20220916174608.csv')

print('Importing file as dataframe: ')
df = pd.read_csv(file_path, delimiter=",", header=1, encoding='latin-1')
print('Import complete.')

df_subset = df[['MAC','SSID','AuthMode','CurrentLatitude','CurrentLongitude']].dropna()

mean_lat = df['CurrentLatitude'].mean()
mean_long = df['CurrentLongitude'].mean()

map = folium.Map(location=[mean_lat,mean_long], zoom_start=17.5)

for i in range(0,len(df_subset),10):
    folium.Marker(
        location = [df_subset.iloc[i]['CurrentLatitude'],df_subset.iloc[i]['CurrentLongitude']],
        popup = df_subset.iloc[i]['SSID']
    ).add_to(map)

display(map)
```
